```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Task 1: Identify the Titanic dataset from the Seaborn library.
df = sns.load_dataset("titanic")

# Task 2: Find the number of male and female passengers in the Titanic dataset.
print(df["sex"].value_counts())
```

```python
# Task 3: Find the number of unique values for each column.
print(df.nunique())
```

```python
# Task 4: Find the number of unique values of the variable pclass.
print(df["pclass"].nunique())
```

```python
# Task 5: Find the number of unique values of pclass and parch variables.
print(df[["pclass", "parch"]].nunique())
```

```python
# Task 6: Check the type of the embarked variable. Change its type to category and check again.
print(df["embarked"].dtypes)
df["embarked"] = df["embarked"].astype("category")
print(df["embarked"].dtypes)
```

```python
# Task 7: Show all the sages of those with embarked value C.
df[df["embarked"]=="C"]
```

```python
# Task 8: Show all the sages of those whose embarked value is not S.
df[df["embarked"]!="S"]
```

```python
# Task 9: Show all information for female passengers younger than 30 years old.
df[(df["sex"]=="female") & (df["age"] < 30)]
```

```python
# Task 10: Show the information of passengers whose Fare is over 500 or 70 years old.
df[(df["fare"] > 500) | (df["age"] > 70)]
```

```python
# Task 11: Find the sum of the null values in each variable.
df.isnull().sum()
df.isna().sum()

df.isnull()
```

```python
# Task 12: Extract the variable who from the dataframe.
df.drop("who", axis=1, inplace=True)
```

```python
# Task 13: Fill the empty values in the deck variable with the most repeated value (mode) of the deck variable.
df["deck"] = df["deck"].fillna(df["deck"].mode()[0], inplace=True)
df["deck"]
```

```python
# Task 14: Fill in the blank values in the age variable with the median of the age variable.
df["age"] = df["age"].fillna(df["age"].median())
df["age"]
```

```python
# Task 15: Find the sum, count, mean values of the pclass and gender variables of the survived variable.
df.groupby(["pclass", "sex"]).agg({"survived": ["sum", "count", "mean"]})
```

```python
# Task 16: Write a function that returns 1 for those under 30 and 0 for those above or equal to 30.
# Using the function you wrote, create a variable named age_flag in the titanic data set. (use apply and lambda constructs)
df["age_flag"] = df["age"].apply(lambda x: 1 if x < 30 else 0)
df.head()
```

```python
# Task 17: Define the Tips dataset from the Seaborn library.
tips = sns.load_dataset("tips")
tips.head(10)
```

```python
# Task 18: Find the sum, min, max and average of the total_bill values according to the categories (Dinner, Lunch) of the Time variable.
tips.groupby("time").agg({"total_bill": ["sum","min", "max", "mean"]})
```

```python
# Task 19: Find the sum, min, max and average of total_bill values by days and time.
tips.groupby(["day","time"]).agg({"total_bill": ["sum","min", "max", "mean"]})
```

```python
# Task 20: Find the sum, min, max and average of the total_bill and type values of the lunch time and female customers according to day.
filtered = tips[(tips["time"]=="Lunch") & (tips["sex"]=="Female")]
filtered.groupby("day").agg({"total_bill": ["sum", "min", "max", "mean"],
                             "tip": ["sum", "min", "max", "mean"]})
```

```python
# Task 21: What is the average of orders with size less than 3 and total_bill greater than 10? (use loc)
filtered2 = tips.loc[(tips['size'] < 3) & (tips['total_bill'] > 10)]
filtered2["total_bill"].mean()
```

```python
# Task 22: Create a new variable called total_bill_tip_sum. Let it give the sum of the total bill and tip paid by each customer.
tips["total_bill_tip_sum"] = tips['total_bill'] + tips['tip']
tips
```

```python
# Task 23: Sort according to the total_bill_tip_sum variable from largest to smallest and assign the first 30 people to a new dataframe.
sorted_tips = tips.sort_values("total_bill_tip_sum", ascending=False)
df_top30 = sorted_tips.head(30)
df_top30
```
