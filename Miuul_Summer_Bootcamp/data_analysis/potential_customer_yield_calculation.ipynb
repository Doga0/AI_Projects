```python
# Q8: What are the PRICE averages by country?
df.groupby("COUNTRY").agg({"PRICE": ["mean"]})
```

```python
# Q9: What are the PRICE averages according to SOURCEs?
df.groupby("SOURCE").agg({"PRICE": ["mean"]})
```

```python
# Q10: What are the PRICE averages in the COUNTRY-SOURCE breakdown?
df.groupby(["COUNTRY","SOURCE"]).agg({"PRICE": ["mean"]})
```

```python
##### Task 2: What are the average earnings in breakdown of COUNTRY, SOURCE, SEX, AGE? #####
agg_df = df.groupby(["COUNTRY","SOURCE","SEX", "AGE"]).agg({"PRICE": "mean"})
```

```python
##### Task 3: Sort the output according to PRICE. #####
agg_df.sort_values("PRICE", ascending=False)
```

```python
##### Task 4: Convert the names in the index to variable names. #####
agg_df = agg_df.reset_index()
agg_df
```

```python
##### Task 5: Convert age variable to categorical variable and add it to agg_df. #####
bins = [0, 18, 23, 30, 40, agg_df["AGE"].max()]
labels = ["0_18", "19_23", "24_30", "31_40", "41"+str(agg_df["AGE"].max())]
agg_df["AGE_CAT"] = pd.cut(agg_df["AGE"], bins=bins, labels=labels)
agg_df
```

```python
##### Task 6: Identify new level-based customers (personas). #####
agg_df["CUSTOMERS_LEVEL_BASED"] = ["_".join(i).upper() for i in agg_df.drop(["PRICE", "AGE"], axis=1).values]
agg_df = agg_df[["CUSTOMERS_LEVEL_BASED", "PRICE"]].groupby("CUSTOMERS_LEVEL_BASED")["PRICE"].mean().reset_index()
agg_df
```

```python
##### Task 7: Segment new customers (personas). #####
agg_df["SEGMENT"] = pd.qcut(agg_df["PRICE"], q=4, labels=["D", "C", "B", "A"])
agg_df.groupby("SEGMENT").agg({"PRICE": ["mean", "max", "min"]}).reset_index()
```

```python
##### Task 8: Categorize new customers and estimate how much revenue they can generate. #####
new_user = "TUR_ANDROID_FEMALE_31_40"
agg_df[agg_df["CUSTOMERS_LEVEL_BASED"] == new_user]

new_user2 = "FRA_IOS_FEMALE_31_40"
agg_df[agg_df["CUSTOMERS_LEVEL_BASED"] == new_user2]

df[["PRICE", "AGE"]].corr()
```
